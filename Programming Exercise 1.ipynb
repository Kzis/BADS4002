## 1.Integer Operators

One way to determine whether an integer is even is to divide the number by two and
check the remainder. Write a three-line program that asks the user to enter a number,
converts the input to an integer, and print a zero if the number is even and a one if the
number is odd.

```python
def main():
    number = getNumber()
    isEven(int(number))

def getNumber():
    number = input("Enter number : ")
    return number
    
def isEven(number):
    if number%2 == 0:
        print("0 : Even")
    else:
        print("1 : Odd")

if __name__ == "__main__":
    main()
```

## 4.Total Purchase

A customer in a store is purchasing five items. Write a program that asks for the price
of each item, and then displays the subtotal of the sale, the amount of sales tax, and
the total. Assume the sales tax is 7 percent.

```python
def main():
    price_of_item = get_price_of_item()
    tax_of_price = get_tax_of_price(price_of_item)
    total_amount = cal_total_amount(price_of_item,tax_of_price)
    show_detail(price_of_item,tax_of_price,total_amount)

def get_price_of_item():
    list_price = list()
    for i in range(1,6):
        price = input("Enter price of item {} : ".format(i))
        list_price.append(float(price))
    return list_price
    
def get_tax_of_price(price_of_item):
    list_tax = list()
    for i in price_of_item:
        tax = i * 0.07
        list_tax.append(round(tax, 2))
    return list_tax
    
def cal_total_amount(price_of_item,tax_of_price):
    list_total = list()
    for i in range(0,5):
        total = price_of_item[i] + tax_of_price[i]
        list_total.append(round(total, 2))
    return list_total

def show_detail(price_of_item,tax_of_price,total_amount):
     for i in range(0,5):
         print("Item {} , price is {} & tax is {} : Total = {}".format(i+1,price_of_item[i],tax_of_price[i],total_amount[i]))

if __name__ == "__main__":
    main()
```

## 5.Tip, Tax, and Total

Write a program that calculates the total amount of a meal purchased at a restaurant.
The program should ask the user to enter the charge for the food, and then calculate
the amount of a 18 percent tip and 7 percent sales tax. Display each of these amounts
and the total.

```python
def main():
    total_purchase = get_total_purchase("Enter total amount of a meal purchased :")
    cal_other = get_other(total_purchase)
    show_details(total_purchase , cal_other)
    
def show_details(purchase , cal_other):
    total = purchase + cal_other[1] + cal_other[0]
    print("Purchase : {} , Tip : {} , Tax : {} , Total : {} ".format(purchase, cal_other[1], cal_other[0],total))
   
def get_total_purchase(text):
    total_purchase = input(text)
    return float(total_purchase)
    
def get_tax(total_purchase):
    return round(total_purchase * 0.07,2)
    
def get_tip(total_purchase):
    return round( total_purchase * 0.18,2)
    
def get_other(total_purchase):
    tax = get_tax(total_purchase)
    tip = get_tip(total_purchase)
    other = (tax,tip)
    return other

if __name__ == "__main__":
    main()
```

## 6.Circle Measurements

Write a program that asks the user to enter the radius of a circle. The program should
calculate and display the area and circumference of the circle.

```python
import math

def main():
    radius = get_radius("Enter radius of circle :")
    circle = get_detail_circle(radius)
    show_details(radius , circle)
    
def show_details(radius , circle):
    print("Radius : {} , Area : {} , Circumference : {}".format(radius , circle[0], circle[1]))

def get_detail_circle(r):
    pi = math.pi
    area = pi * (r ** 2)
    circumference = 2 * pi *r
    details = area , circumference
    return details

def get_radius(text):
    return float(input(text))

if __name__ == "__main__":
    main()
```

## 7.Ingredient Adjuster
A cookie recipe calls for the following ingredients:
- 1.5 cups of sugar
- 1 cup of butter
- 2.75 cups of flour

The recipe produces 48 cookies with this amount of the ingredients. Write a program
that asks the user how many cookies he or she wants to make, and then displays the
number of cups of each ingredient needed for the specified number of cookies.

```python
def main():
    number_of_cookies = get_cookies("Enter  number of cookies :")
    ingredients = get_ingredients(number_of_cookies)
    show_details(number_of_cookies , ingredients)
    
def show_details(number_of_cookies , ingredients):
    print("Cookies {} piece need sugar {} cup & butter {} cup & flour {} cup".format(number_of_cookies,ingredients[0],ingredients[1],ingredients[2]))
    
def get_cookies(text):
    return int(input(text))
    
def get_ingredients(n):
    sugar_per_piece = 1.5/48
    butter_per_piece = 1/48
    flour_per_piece = 2.75/48
    ingredients_need = (n*sugar_per_piece, n*butter_per_piece, n*flour_per_piece)
    return ingredients_need

if __name__ == "__main__":
    main()
```

## 9.Stock Transaction Program
Last month Joe purchased some stock in Acme Software, Inc. Here are the details of
the purchase:
- The number of shares that Joe purchased was 2,000.
- When Joe purchased the stock, he paid 40.00 per share.
- Joe paid his stockbroker a commission that amounted to 3 percent of the amount he paid for the stock.

Two weeks later Joe sold the stock. Here are the details of the sale:
- The number of shares that Joe sold was 2,000.
- He sold the stock for 42.75 per share.
- He paid his stockbroker another commission that amounted to 3 percent of the amount he received for the stock.

Write a program that displays the following information:
- The amount of money Joe paid for the stock.
- The amount of commission Joe paid his broker when he bought the stock.
- The amount of which Joe sold the stock.
- The amount of commission Joe paid his broker when he sold the stock.
- Display the amount of money that Joe had left when he sold the stock and paid his
broker (both times). If this amount is positive, then Joe made a profit. If the
amount is negative, then Joe lost money.

```python
def main():
    joe_purchased = [(2000,40.00,3,'buy'),(2000,42.75,3,'sold')]
    joe_details = cal_money(joe_purchased)
    print_result(joe_purchased,joe_details)
    
def show_details(result):
    print("The amount of money Joe paid for the stock : {}".format(result["purchased_stock"]))
    print("The amount of commission Joe paid his broker when he bought the stock : {}".format(result["first_commission"]))
    print("The amount of which Joe sold the stock : {}".format(result["sold_stock"]))
    print("The amount of commission Joe paid his broker when he sold the stock : {}".format(result["second_commission"]))
    print("----------")
    print("The amount of commission Joe paid his broker first time is {} , second time is {} , total is {}".format(result["first_commission"], result["second_commission"] , result["total_commission"]))
    print("Joe paid is {} , Joe recieve is {} so he {}".format(result["purchased_total"],result["sold_stock"],result["result"]))
    
def print_result(purchased , details):
    purchased_stock = details[0][0]
    first_commission = details[0][1]
    sold_stock = details[1][0]
    second_commission = details[1][1]
    total_commission = first_commission + second_commission
    purchased_total = purchased_stock + total_commission
    result = " Joe made a profit" if sold_stock > purchased_total else " Joe lost money"
    
    obj_result = {
        "purchased_stock": purchased_stock,
        "first_commission": first_commission,
        "sold_stock":sold_stock,
        "second_commission":  second_commission,
        "total_commission": total_commission,
        "purchased_total": purchased_total,
        "result": result,
    }
    
    show_details(obj_result)
    
    
def cal_money(details):
    stock_details = list()
    for item in details:
        money = item[0] * item[1]
        commission = (money * 0.03)
        total = money + commission if item[3] == 'buy' else money - commission
        stock_details.append((money , commission , total))
    return stock_details

if __name__ == "__main__":
    main()
```
